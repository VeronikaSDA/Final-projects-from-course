In this solution I tried to load all pages from the blog, then I wanted go back to up of the page and load title of article, introduction of article, then open article and load its text, then go back to firt page and open next article and so on. Unfortunately, It work only till article, where are 2 videos and it can't return back to first page. 

```python
import requests
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.common import ElementClickInterceptedException, NoSuchElementException
from selenium.webdriver.common.by import By
from time import sleep
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC



def stazeni_clanku():
    seznam = []
    service = Service()
    browser = webdriver.Chrome(service=service)
    browser.get("https://digitalniarchitekti.cz/blog/")

#setting of cookies
    ochrana_nastaveni = False
    if not ochrana_nastaveni:
        ochrana = browser.find_elements(By.ID, "onetrust-accept-btn-handler")
        if len(ochrana) > 0:
            print("Nastavení ochrany nalezeno")
            ochrana[0].click()   
            ochrana_nastaveni = True
        else:
            print("Nastavení ochrany nenalezeno")

#loading all pages/articles
    try:
        while True:
            next_page = browser.find_element(By.XPATH, "//span[@class='ast-load-more active']")
            next_page.click()
            sleep(5)
    except ElementClickInterceptedException:
        print("Tlačítko nebylo možné stisknout.")
        sleep(5)
    except NoSuchElementException:
        print("Tlačítko se již nezobrazuje, jsi na konci.")
        sleep(5)
    print(f"Vsechny stranky byly nacteny.")

#going back to the top of page
    nahoru = browser.find_element(By.ID, "ast-scroll-top")
    nahoru.click()

    container = browser.find_element(By.ID, "main")
    articles = container.find_elements(By.CSS_SELECTOR, "article")

#opening of article, loading information and saving it, going back to the first page
    for article in articles:
        uvodnik = article.find_element(By.CSS_SELECTOR, ".ast-excerpt-container.ast-blog-single-element")
        uvodnik = uvodnik.find_element(By.CSS_SELECTOR, "p").text
        print(uvodnik)

        tag = article.find_element(By.TAG_NAME,"a")
        url = tag.get_attribute("href")
        print(url)
        tag.click()
        sleep(5)

        browser.get(url)
        nadpis  = browser.find_element(By.CSS_SELECTOR, ".entry-title").text
        print (nadpis)
           
        clanek =  browser.find_element(By.CSS_SELECTOR, ".entry-content.clear")
        clanek_obsah = clanek.find_elements(By.TAG_NAME, "p")
        clanek_text = ""
        for part in clanek_obsah:
            clanek_text += part.text + "\n"
        print(clanek_text)
        sleep(5)
        
        slovnik = {"nadpis" : nadpis,
               "uvodnik" : uvodnik,
               "clanek" : clanek_text}
        seznam.append(slovnik)
        
        sleep(5)

        browser.back()
        sleep(5)

    browser.quit()
    return seznam

vysledek = stazeni_clanku()
print(vysledek)  
    
        
```

```python
#saving into the CSV file
df = pd.DataFrame(vysledek)
df

try:
    soubor = df.to_csv("Články_blog_DA2.csv", index = False, encoding = "utf-8")
    print("Dokument byl vytvořen.")
except:
    print("Dokument nebyl vytvořen!")
```
