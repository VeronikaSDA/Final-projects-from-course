```python
import requests
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.common import ElementClickInterceptedException, NoSuchElementException
from selenium.webdriver.common.by import By
from time import sleep
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC



def stazeni_clanku():
    seznam = []
    service = Service()
    browser = webdriver.Chrome(service=service)
    browser.get("https://digitalniarchitekti.cz/blog/")

# setting of cookies
    ochrana_nastaveni = False
    if not ochrana_nastaveni:
        ochrana = browser.find_elements(By.ID, "onetrust-accept-btn-handler")
        if len(ochrana) > 0:
            print("Nastavení ochrany nalezeno")
            ochrana[0].click()   
            ochrana_nastaveni = True
        else:
            print("Nastavení ochrany nenalezeno")

#loading of area with articles
    container = browser.find_element(By.ID, "main")
    articles = container.find_element(By.CSS_SELECTOR, "article")

#opening of first article
    tag = articles.find_element(By.TAG_NAME,"a")
    url = tag.get_attribute("href")
    print(url)
    tag.click()
    sleep(5)
    browser.get(url)

#loading of wanted information, saving it to dictionary and list and loading next article
    try:
        while True:
            nadpis  = browser.find_element(By.CSS_SELECTOR, ".entry-title").text
            print (nadpis)

            uvodnik = browser.find_element(By.CSS_SELECTOR, ".entry-content.clear")
            uvodnik_obsah = uvodnik.find_element(By.CSS_SELECTOR, "p").text
            print(uvodnik_obsah)

            clanek =  browser.find_element(By.CSS_SELECTOR, ".entry-content.clear")
            clanek_obsah = clanek.find_elements(By.TAG_NAME, "p")
            clanek_obsah = [x.text for x in clanek_obsah]
            clanek_obsah = "\n".join(clanek_obsah)
            print(clanek_obsah)
        
            slovnik = {"nadpis" : nadpis,
                "uvodnik" : uvodnik_obsah,
                "clanek" : clanek_obsah}
            seznam.append(slovnik)
        
            sleep(5)

            next = browser.find_element(By.CSS_SELECTOR, ".nav-previous")
            web = next.find_element(By.TAG_NAME,"a")
            next = web.get_attribute("href")   
            web.click()
            browser.get(next)

    except ElementClickInterceptedException:
        print("Tlacitko nebylo mozne stisknout.")
        sleep(5)
    except NoSuchElementException:
        print("Tlacitko se jiz nezobrazuje, jsi na konci.")
        sleep(5)
    print("Vsechny clanky byly nacteny.")
        
            
    browser.quit()
    return seznam

vysledek = stazeni_clanku()
print(vysledek)  
    
        
```

```python
#printing number of saved articles
print(f"Bylo stazeno {len(vysledek)} clanku.")
```

```python
#making of CSV file
import datetime

df = pd.DataFrame(vysledek)
df

today = datetime.date.today()
try:
    soubor = df.to_csv(f"Články_blog_DA_{today}.csv", index = False, encoding = "utf-8")
    print("Dokument byl vytvořen.")
except:
    print("Dokument nebyl vytvořen!")

df
```
