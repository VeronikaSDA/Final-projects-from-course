```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st

data = pd.read_csv("job_offers_2024-06-27_cleaned.csv",
                       encoding = "utf-8",
                       sep = ",",
                       decimal = ".")


#count of offers per position and make a bar chart
print(data["kategorie"].value_counts())
count = data["kategorie"].value_counts()

pos_name = count.index.values 

x = np.arange(len(pos_name))
y = count
plt.figure(figsize=(10,5))
plt.bar(x, y, color='maroon')
 
plt.xticks(x, pos_name)
plt.yticks(np.arange(0,max(y)+1, 1))
 
plt.title('Number of offers per position')
 
plt.ylabel('Počet nabídek')
plt.xlabel('Pozice')
 
plt.minorticks_on()
plt.legend()
 

plt.grid(which='major', linestyle='-', linewidth='0.2', color='black')

for i in range(len(pos_name)):
    plt.text(x[i], y[i] / 2, f"{y[i]:.0f}", ha = 'center', fontsize = 9, fontweight='bold')
    
plt.tight_layout

plt.savefig("Number_of_offers_per_position.jpg")
plt.show()

```

```python
#Average salary per position, bar chart

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st

data = pd.read_csv("job_offers_2024-06-27_cleaned.csv",
                       encoding = "utf-8",
                       sep = ",",
                       decimal = ".")

data.head()
data_avg = data.groupby("kategorie")[["dolni hranice", "horni hranice", "mzda prumer"]].mean()
data_avg
pos_name = data_avg.index.values
pos_name
y1 = data_avg["dolni hranice"]
y2 = data_avg["horni hranice"]
y3 = data_avg["mzda prumer"]
width = 0.2

x = np.arange(len(pos_name))
plt.figure(figsize=(12,6))
plt.bar(x - width, y1, width, label = "Minimum", color='cyan')
plt.bar(x, y2, width, label = "Maximum", color='salmon')
plt.bar(x + width, y3, width, label = "Total average", color='magenta')
 
plt.xticks(x, pos_name)
plt.yticks(np.arange(0,max(y2) + 500, 1000))
  
plt.title('Average salary by position')
 
plt.ylabel('Average salary')
plt.xlabel('Pozice')


plt.legend()
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.2', color='black')

plt.tight_layout

for i in range(len(pos_name)):
    plt.text(x[i] - width, y1[i] + 200, f"{y1[i]:.0f}", ha = 'center', fontsize = 9, fontweight='bold')
    plt.text(x[i], y2[i] + 200, f"{y2[i]:.0f}", ha = 'center', va = 'center', fontsize = 9, fontweight='bold')
    plt.text(x[i] + width, y3[i] + 200, f"{y3[i]:.0f}", ha = 'center', va = 'center', fontsize = 9, fontweight='bold')


plt.savefig("Average_salary_by_position.png")
plt.show()

  
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st

data = pd.read_csv("job_offers_2024-06-27_cleaned.csv",
                       encoding = "utf-8",
                       sep = ",",
                       decimal = ".")


#offer by city, bar chart

print(data["lokalita"].value_counts())
count_city = data["lokalita"].value_counts()
jmena = count_city.index.values

x = np.arange(len(jmena))
y = count_city

plt.figure(figsize=(10,5))
plt.bar(x, y, 0.4, color='teal')
 
plt.xticks(x, jmena, rotation = 45, fontsize = 8)
plt.yticks(np.arange(0,max(y)+1, 2))
 
plt.title('Number of offers per city')
 
plt.ylabel('Počet nabídek')
plt.xlabel('Města')
 
plt.minorticks_on()
plt.legend()
 

plt.grid(which='major', linestyle='-', linewidth='0.2', color='black')

for i in range(len(jmena)):
    plt.text(x[i], y[i] + 0.5, f"{y[i]:.0f}", ha = 'center', fontsize = 9, fontweight='bold')
    
plt.tight_layout

plt.savefig("Number_of_offers_per_city.png")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st

data = pd.read_csv("job_offers_2024-06-27_cleaned.csv",
                       encoding = "utf-8",
                       sep = ",",
                       decimal = ".")

#highest salary

max = data["horni hranice"].max()
max

city = data.groupby("lokalita")["horni hranice"].max().sort_values(ascending = False)
print(city)
city_name = city.index.values
x = np.arange(len(city_name))
y = city.values
y_top = y.max()

plt.figure(figsize=(13,7))
plt.bar(x, y, 0.3, color = "lightcoral")

plt.xticks(x, city_name, rotation = 45, fontsize = 8)
plt.yticks(np.arange(0, y_top + 1000, 2000))
 
plt.title('Highest salary per city')
 
plt.ylabel('Top level salary')
plt.xlabel('Města')

plt.grid(which='major', linestyle='-', linewidth='0.2', color='black')

for i in range(len(city_name)):
    plt.text(x[i], y[i] + 0.5, f"{y[i]:.0f}", ha = 'center', fontsize = 9, fontweight='bold')
    
plt.tight_layout

plt.savefig('Highest_salary_per_city.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st

data = pd.read_csv("job_offers_2024-06-27_cleaned.csv",
                       encoding = "utf-8",
                       sep = ",",
                       decimal = ".")

#senior vs junior postions

data = data.groupby("senior pozice")["mzda prumer"].mean()
data
name = ["regular position", "senior position"]

x = np.arange(len(name))
y = data.values
y_top = y.max()
plt.figure(figsize=(10,5))
plt.bar(x, y, 0.4, color='mediumpurple')

 
plt.xticks(x, name)
plt.yticks(np.arange(0, y_top + 1000, 2000))
  
plt.title('Average salary by seniority')
 
plt.ylabel('Average salary')
plt.xlabel('Typ pozice')


plt.minorticks_on()
plt.tight_layout
plt.grid(which='major', linestyle='-', linewidth='0.2', color='black')


for i in range(len(name)):
    plt.text(x[i], y[i] + 200, f"{y[i]:.0f}", ha = 'center', va = 'center', fontsize = 9, fontweight='bold')
    
plt.savefig('Average salary by seniority.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st

#Where is possible to hire a team of data analyst, data engineer and data scientist for lowest cost?

data = pd.read_csv("job_offers_2024-06-27_cleaned.csv",
                       encoding = "utf-8",
                       sep = ",",
                       decimal = ".")

DA_min = data.query("kategorie == 'data analyst'")
DA = DA_min["dolni hranice"].min()

DE = data.query("kategorie == 'data engineer'")
DE_min = DE["dolni hranice"].min()

DS = data.query("kategorie == 'data scientist'")
DS_min = DS["dolni hranice"].min()

print(f"""Spodni hranice pro Data Analyst: {DA},
spodní hranice pro Data Engineer: {DE_min},
spodní hranice pro Data Scientist: {DS_min}""")


hire_cost_min = DA + DE_min + DS_min
hire_cost_min
print(f"Celý tým lze najmout nejméně za {hire_cost_min} PLN.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st

#number of positions by city, where is possible to higher whole team, what cost for whole team in Warszawa or Remote

data = pd.read_csv("job_offers_2024-06-27_cleaned.csv",
                       encoding = "utf-8",
                       sep = ",",
                       decimal = ".")


data_group = data.groupby("kategorie")["lokalita"].value_counts()
print(data_group)
lokality = data_group.index.get_level_values(1).unique()
print(f"""
Dostupné lokality jsou: {lokality[0]},{lokality[1]}, {lokality[2]}, {lokality[3]}, {lokality[4]}, {lokality[5]}.
      """)

print("""Celý tým je možné najmout v lokalitě Remote, Warszawa.
      """)

remote = data.query("lokalita == 'remote'")
remote = remote.groupby("kategorie")["dolni hranice"].min()
soucet = remote.sum()
print(remote)
print(f"""
Tým v lokalitě Remote je možné najmout za {soucet} PLN.
""")

Warszawa = data.query("lokalita == 'warszawa'")
Warszawa = Warszawa.groupby("kategorie")["dolni hranice"].min()
soucet_2 = Warszawa.sum()
print(Warszawa)
print(f"""
Tým v lokalitě Warszawa je možné najmout za {soucet_2} PLN.
""")

if soucet < soucet_2:
    print("Levnější tým vychází v lokalitě Remote.")
else:
    print("Levnější tým vychází v lokalitě Warszawa.")

```
